```python
import ar2gas as gas
import pygeostat as gs
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = gs.DataFile(flname='dataset.csv', x='x', y='y')
```

```python
gs.locmap(data, x='x', y='y', figsize=(10,10))
```

```python
codes = data.data.code.unique()
fig, axes = plt.subplots(1, 3, figsize=(15,5))
for idx, c in enumerate(codes):
    data['ind_{}'.format(c)] = np.where((data['code'] == c) == True, 1, 0)
    vario = gs.Variogram(data, 'ind_{}'.format(c), ndim=2, ndir=1, omnihz=True)
    vario.settols(nlags=12, lagdist=50, lagtol=25)
    vario.varcalc()
    vario.fitmodel(sill=np.var(data['ind_{}'.format(c)]))
    print('\n Var model: {}\n'.format(vario.model))
    ax = vario.plot(titles='code {}'.format(int(c)), model=True, sill=np.var(data['ind_{}'.format(c)]), ylim=[0.0,0.3], axes=axes[idx])
    plt.tight_layout()
```
